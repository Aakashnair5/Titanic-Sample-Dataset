```python
#importing dependencies
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.preprocessing import LabelEncoder
print("Imported Successfully All")
```

```python
#bringing in the data
df=pd.read_csv("train[1].csv")
df.head()
le=LabelEncoder()
df.head()
df['nFare']=le.fit_transform(df['Fare'])
df.head()
```

```python
inputs=df.drop('nFare',axis='columns')
target=df['nFare']

target.head()
inputs.head()

```

```python
inputs.drop('Fare',axis=1,inplace=True)
```

```python
inputs['NewName']=le.fit_transform(inputs['Name'])
inputs['Gender']= le.fit_transform(inputs['Sex'])

inputs['Token']= le.fit_transform(inputs['Ticket'])
inputs['Nage']= le.fit_transform(inputs['Age'])

inputs.head()



```
